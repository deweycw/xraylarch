```python
import numpy as np
import larch
from larch.xafs import pre_edge, autobk
from larch.io import read_ascii
from larch.plot.bokeh_xafsplots import plot_mu, plot_bkg, plot_chik

cu = read_ascii('../xafsdata/cu_metal_rt.xdi')
cu.mu = -np.log(cu.itrans/cu.i0)
pre_edge(cu)
autobk(cu, rbkg=1, kw=2)
print(f'{cu.e0=}')
```

```python
fig = plot_mu(cu, with_deriv=False, show_e0=True, emin=8900, emax=9200)
```

```python
fig = plot_bkg(cu, emin=8900, emax=9600)
```

```python
plot_chik(cu, kweight=2)
```
